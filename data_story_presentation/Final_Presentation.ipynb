Earlier this year, [a 30-year-old](https://www.nbcnewyork.com/news/local/woman-beaten-nearly-raped-on-brooklyn-street-in-brutal-attack/2310269/) woman walking back to her home in Brooklyn in the early morning was suddenly attacked from behind and nearly raped after falling to the ground.  This past May, [a 64-year-old](https://www.nbcnewyork.com/news/local/rape-victim-dies-9-days-after-ambush-attack-on-manhattan-sidewalk/2448782/) woman walking in Manhattan a little before midnight was slammed against a wall, raped on the floor, and later died from her injuries. 

It’s no surprise that women often express the fear of walking alone at night. Sexual assault and rape affect millions of people each year in the United States. [The CDC](https://www.cdc.gov/violenceprevention/datasources/nisvs/2015NISVSdatabrief.html) states that 1 in 5 women and 1 in 14 men will experience attempted or completed rape in their lifetime. In most places, sexual assault is generally defined as subjecting another person to sexual contact without the latter's consent. This includes groping strangers on the street, incest, child molestation, and rape, but also includes actions like forcing your partner to have sex or drunken attempts to move on a friend without their consent. In fact, [51.1% of female rape victims](https://www.nsvrc.org/statistics) reported being raped by an intimate partner. Fortunately, we’ve seen more sexual assault awareness efforts in recent years, beginning with the Anti-Rape movement of the 1970s to the #MeToo Movement in 2017. With unprecedented increases in sexual assault awareness and prevention efforts, the expectation is that the number of sexual assault cases would decrease.

Well, looking at the crime data from 2013 to the end of last year, that doesn’t seem to be the case.




*Chart by Jean Hsu*, Source: [Bureau of Justice Statistics' Criminal Victimization Series](https://www.bjs.gov/index.cfm?ty=tp&tid=317)


In the chart above, the blue line indicates the number of sexual assault victims in the U.S. from 2013-2019. Just last year in 2019, there were 459,310 victims of sexual assault, a 53% increase compared to the number of victims in 2013. Additionally, there was a drastic increase in victims in 2018, reaching up to 734,630 victims of sexual assault within just one year, more than the victims in 2013 and 2014 combined. 

Considering the widespread national exposure of the sexual-abuse allegations against Harvey Weinstein in 2017 and the viral #MeToo movement that followed, you’d expect numbers to decrease or at least stay the same in 2018. Instead, the number of victims peaked in 2018 and increased by 86% since the previous year. While it’s evident that sexual assault is still a major concern in the U.S., it’s worrisome that widespread awareness can have little to no effect on the issue. 

### Victims of Sexual Assault

Women, younger women especially, experience sexual violence at higher rates than men. An estimated [17.7 million women](https://www.rainn.org/statistics/victims-sexual-violence) in the U.S. had experienced attempted or completed rape since 1998. [Women ages 16-19](https://www.rainn.org/statistics/victims-sexual-violence) are also 4 times more likely than the general population to be victims of sexual assault. 

After looking closely at the sexual assault data for specifically New York City, I found that these national trends remained true at a local level. The chart below on the left shows the number of sexual assault cases in NYC separated by victim age groups. It's visible that a majority of sexual assault victims in NYC since 2006 have been under the age of 18. Additionally, out of the victims under the age of 18, I found that about 78% of them were female. 

The chart below on the right shows the sex of sexual assault suspects and victims in NYC since 2006. The orange represents the number of male-identifying individuals and the blue represents the number of female-identifying individuals. It's apparent that among the known suspects, an overwhelming majority have been male-identifying individuals, while among the victims, an overwhelming majority have been female-identifying individuals. Just like the national statistics, younger women are more at risk of being victims of sexual assault in NYC.

### Underreporting of Sexual Assault 

Another major issue with sexual assault in the U.S. is the chronic underreporting of cases. The chart below shows the percentage of sexual assault cases in the U.S. not reported to the police every year from 2013 to 2019.




*Chart by Jean Hsu*, Source: [Bureau of Justice Statistics' Criminal Victimization Series](https://www.bjs.gov/index.cfm?ty=tp&tid=317)


For every year since 2013, more than half of all sexual assault cases were not reported to the police. Together, about 68% of all cases since 2013 remain unreported. In 2018, the number of unreported sexual assault reached a high point of 75.1%. If this percentage was applied to the number of sexual assault victims in 2018, that would mean more than 550,000 victims did not report their case to the police that year. 

### Sexual assault is a serious offense that can cause victims' mental health difficulties for years to come. So, **why aren’t they reported?**

Based on the [Department of Justice reports](https://www.rainn.org/statistics/criminal-justice-system), one reason sexual assault victims choose not to report is due to the fear of retaliation from the perpetrator, as well as social recrimination and stigmatization. Some victims even believe that their matter isn’t important enough to report. Another important reason is that victims often believe that the police or criminal justice system would not or could not do anything to help their situation.

To learn more about this, I dug deeper into the sexual assault reports and arrests in New York City.

After looking at the data from New York City's sexual assault reports and arrests from 2006, I believe it's completely valid for victims to feel like the justice system would not be on their side. The chart above on the left shows the number of sexual assault reports (blue) and sexual assault arrests (orange) in NYC since 2006. As you can see, the number of sexual assault arrests made by the NYPD is significantly lower than the number of sexual assault cases reported to them each year. Even when reported cases peaked in 2018, the number of arrests did not change much from previous years. 

Furthermore, the chart above on the right shows the ratio of sexual assault reports to sexual assault arrests in NYC. The relatively high ratio of reports to arrests shows that there are many more sexual assault complaints reported than the number of arrests made each year. For example, the ratio in 2017 and 2018 were over 6, which means that for about every 6 sexual assault cases reported, there was only one sexual assault arrest made. In addition, the ratio of sexual assault reports to arrests in NYC has gradually increased since 2006 and the ratio has been over 4.4 since 2011. 

Also, according to RAINN (Rape, Abuse and Incest National Network), out of the 31% of sexual assault cases reported to the police in the U.S., [only 5.7% end in arrest and 0.7% result in a felony conviction.](https://www.washingtonpost.com/business/2018/10/06/less-than-percent-rapes-lead-felony-convictions-least-percent-victims-face-emotional-physical-consequences/) Looking at this national and local data, it’s understandable that many survivors of sexual assault would believe that reporting their cases to the police would not lead to any meaningful consequences for the perpetrators.

Despite many sexual assault survivors choosing not to rely on the criminal justice system known for its low conviction rates, slow processing of rape kits, and reputation of retraumatizing victims, they have found other ways to take power back from their attackers and heal from their trauma. The Me Too movement, founded by Tarana Burke in 2006 and made viral by Alyssa Milano in 2017, has brought massive awareness to sexual assault and inspired many survivors to stand up and speak about their traumas. This has led to the legal action taken against many powerful men, like former [USA Gymnastics doctor Larry Nassar](https://www.usatoday.com/pages/interactives/larry-nassar-timeline/) and [singer-songwriter R. Kelly](https://time.com/5546990/r-kelly-timeline/),  who have been called out for assaulting multiple women throughout their careers.  Earlier this year as well, multiple personal and university-specific Twitter accounts were made to expose anonymous sexual assault allegations. For example, the Twitter account [“Assaulters at UMich”](https://www.michigandaily.com/section/news/twitter-account-about-sexual-assault-%E2%80%98u%E2%80%99-goes-viral) posted a series of tweets exposing alleged sexual assaulters at the University of Michigan, including  a professor.

Overall, it’s apparent that sexual assault remains a serious problem in the United States. Reported sexual assault cases in the U.S. have not only increased during the past 7 years, but the percentage of sexual assault cases not reported to the police has remained at relatively high levels as well. Furthermore, the sexual assault cases that are reported show that younger women continue to be at overwhelmingly higher risks of sexual assault. The staggering amount of victims that choose not to report also underscores the faulty criminal justice system that needs to be reformed. A system that allows so many sexual perpetrators to walk free further communicates to society that sexual assault can be done without major consequences. To decrease the number of sexual assault cases and protect those at risk, serious reforms must be made to not just our criminal justice system, but also in the way we develop our educational and preventative efforts.
